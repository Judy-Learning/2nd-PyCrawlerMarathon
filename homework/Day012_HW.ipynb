### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
dic = {}
title=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    cls = d.find(class_="tag").text
    title=[d.find_all('a')[-1].text]
    if cls in dic:
        dic[cls].extend(title)
    else:
        dic[cls]=title
for k in dic:
    for v in dic[k]:
        print(k,' ',v)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
for k in dic:
    print(k,' ',len(dic[k]),'筆')
print('\n** 生活類的文章最多.')
```
