## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

# 每隔1秒鐘自動往下滑

import time
from datetime import datetime, timedelta

now = datetime.today()
print('Start_Now: ', now.strftime("%Y/%m/%d %H:%M:%S"))
nowDate = now.strftime("%Y/%m/%d")


for i in range(60):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 30000);")
    
    
# 取得資料，丟到 BeautifulSoup 解析


print('\n今天所有的新聞標題:')

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    NewsDate = d.find(class_="date").text[0:10]
    if nowDate == NewsDate:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();

now = datetime.today()
print('\nEnd_Now: ', now.strftime("%Y/%m/%d %H:%M:%S"))
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")


# 每隔1秒鐘自動往下滑

import time
from datetime import datetime, timedelta

now = datetime.today()
print('Start_Now: ', now.strftime("%Y/%m/%d %H:%M:%S"))

for i in range(10):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 30000);")


# 取得資料，丟到 BeautifulSoup 解析

_2hoursago = now - timedelta(hours=2)
print('\n從', _2hoursago.strftime("%Y/%m/%d %H:%M"),'到現在(兩小時內)的新聞標題:')

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    newsDateTime = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if _2hoursago <= newsDateTime:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();

now = datetime.today()
print('\nEnd_Now: ', now.strftime("%Y/%m/%d %H:%M:%S"))
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list-2020-3-26-0.htm")


# 每隔1秒鐘自動往下滑

import time
from datetime import datetime, timedelta

now = datetime.today()
print('Start_Now: ', now.strftime("%Y/%m/%d %H:%M:%S"))

for i in range(40):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 30000);")


# 取得資料，丟到 BeautifulSoup 解析

from datetime import datetime, timedelta

_3daysago = datetime.today() - timedelta(days=3)
print('\n', _3daysago.strftime("%Y/%m/%d"), '三天前下午三點到五點的新聞')

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    NewsHour = d.find(class_="date").text[-5:-3]
    if NewsHour in ['15','16']:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();

now = datetime.today()
print('\nEnd_Now: ', now.strftime("%Y/%m/%d %H:%M:%S"))
```
