```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://github.com/new', verify = False)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

```python
rs = requests.session()
payload={
    'from':'https://github.com/new',
    'yes': 'yes'
}
res = rs.post('https://github.com/login',verify = False, data = payload)
res = rs.get('https://github.com/new',verify = False)
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text) 
```
